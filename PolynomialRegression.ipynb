```python
import pandas as pd
```

```python
data = {
    "Experience": [1,2,3,4,5],
    "Salary": [30,35,50,70,110]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
X = df[["Experience"]]
y = df[["Salary"]]
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Step 5: Create Polynomial Features (degree = 2)

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
# Step 6: Train Linear Regression on Polynomial Features 

poly_model = LinearRegression()
poly_model.fit(X_poly, y)
```

```python
# Step 7: Check coefficients and intercept

print("Polynomial Coefficients:", poly_model.coef_)
print("Polynomial Intercept:", poly_model.intercept_)
```

```python
# Step 8: make predictions 
y_poly_pred = poly_model.predict(X_poly)

print("Predicted Salary:", y_poly_pred)
```
